## Variable scope

Variable scope in Python determines where a variable can be accessed and how long it exists during program execution. There are two main types: local scope and global scope. Local variables are created inside a function and can only be used within that function; once the function finishes running, they are no longer accessible. In contrast, global variables are defined outside functions and can be accessed throughout the entire program. While global variables are useful for shared information, they must be used carefully because changes in one part of the program can affect other parts in unintended ways. Understanding scope is essential for writing clean, organised, and reliable Python code.

### Best Practices for Global Variables

Global variables should be used carefully and only when necessary. They are most suitable for storing constant values that do not change during program execution, such as configuration settings, mathematical constants, or file paths.

When data needs to be shared across multiple functions, global variables can be used, but it is often better to encapsulate data within functions and provide controlled access to it instead. This improves code safety and maintainability.

For larger programs, splitting code into smaller, more focused modules is recommended. This reduces reliance on global variables and leads to a cleaner, more modular design.

Using global variables wisely helps maintain organised, efficient, and manageable code.


### The LEGB Rule: Python’s Variable Lookup Strategy

When a variable is used, Python searches for its value in a specific order known as the LEGB rule.

- Local (L): Python first looks for the variable inside the current function or local scope.

- Enclosing (E): If not found locally, it searches in enclosing functions in cases of nested functions.

- Global (G): If still not found, it checks the global scope, meaning variables defined at the top level of a module.

- Built-in (B): Finally, Python checks the built-in namespace, which includes predefined functions and objects such as `print()` and `list()`.

This structured search process ensures that Python resolves variables in a predictable way and avoids conflicts between names defined at different levels.

### Nested scope

In more complex Python programs, you might have functions within functions (nested functions). In this case, a variable in the outer function (enclosing scope) is visible to the inner function, even if it's not global.



````python
#Nested scope example

def outer_function():
    outer_var = "Hello from the outer function!"
    def inner_function():
        print(outer_var)  # Accessible from the enclosing scope
    inner_function()
outer_function()
# Two nested functions, the outer one labeled outer_function and the inner one inner_function
````

# 2.  Classes and Object-Oriented Programming (OOP) in Python

Classes are a fundamental concept in object-oriented programming (OOP) in Python. They act as blueprints for creating objects, which are the core building blocks of applications. A useful analogy is that classes are like moulds for cookies: each cookie (object) is unique, but all share the same underlying shape and structure defined by the class. Classes allow you to define both the properties (attributes) and behaviours (methods) that objects will have.


# Understanding Object-Oriented Programming

Imagine building a virtual zoo. In an object-oriented approach, each animal is represented as an object, which is an instance of a class. For example, a `Lion` class may include attributes such as name, species, age, and mane colour, and methods such as `roar()`, `hunt()`, and `sleep()`. A `Monkey` class would have its own specific attributes and methods. Each object is therefore a self-contained representation of a real-world entity with both data and behaviour.


# Key Components of OOP

## Objects

Objects represent real-world entities or abstract concepts in a program. Each object contains its own data and can perform actions defined by its class, making it a self-contained unit.

## Classes

Classes are templates used to create objects. They define shared attributes and methods that all objects of that type will have, ensuring consistency and structure.

## Encapsulation

Encapsulation refers to the practice of keeping an object’s internal data and implementation details hidden from the outside. It allows data and the methods that operate on it to be bundled within a class, protecting the internal state and preventing unintended interference.

## Abstraction

Abstraction simplifies complex systems by exposing only the essential features while hiding internal implementation details. It focuses on what an object does rather than how it does it, making code easier to use and understand.

## Inheritance

Inheritance allows a new class to be created based on an existing class. The new class inherits attributes and methods from the parent class while also allowing additional features or modifications. This promotes code reuse and models hierarchical relationships.

## Polymorphism

Polymorphism allows different objects to be treated in a uniform way as long as they share a common interface. This enables flexibility in how objects are used and improves the adaptability of code.

## Why do we need Classes?

At a basic level, Python programs use variables, functions, and modules. While these tools are effective, larger programs can become difficult to manage when everything is loosely connected. Classes help reduce this complexity by organising related data and behaviour into a single structure.


## Code Organisation

Classes provide a clear structure for organising code. Instead of spreading variables and functions throughout a program, they allow related elements to be grouped together in one place. This improves readability, maintainability, and overall clarity.

For example, in a game with multiple characters, each character may have attributes such as name, health, strength, and speed. Without classes, this can lead to many separate variables like `character1_name` or `character2_health`, which becomes difficult to manage. A `Character` class allows all of these attributes to be neatly contained within a single object.


## Reusability

Once a class is defined, it can be used to create multiple objects without rewriting code. Each object is an instance of the same class but can hold different values.

For example, a `BankAccount` class can be used to create multiple accounts such as `my_account` and `your_account`. Each account will have its own balance and transaction history, while the logic for deposits and withdrawals is written only once inside the class.


## Encapsulation

Encapsulation in Python is based on convention rather than strict enforcement. It is commonly implemented using a single underscore prefix (`_`) to indicate that an attribute is intended for internal use within a class.

Although Python does not prevent access to these attributes from outside the class, the underscore serves as a signal to other developers that the data is part of the internal implementation and should not be modified directly. This helps protect the integrity of the object and supports better code design.

````python
class BankAccount:
    '''
    In this example, the _balance attribute is marked as private. While you could technically access and modify it 
    directly from outside the class (for example., my_account._balance = 1000), doing so would violate the 
    principle of encapsulation. The class provides the deposit() and get_balance() methods as the proper 
    interface for interacting with the account's balance.
    '''
    def __init__(self, balance):
        self._balance = balance  # Private attribute

    def deposit(self, amount):
        self._balance += amount

    def get_balance(self):
        return self._balance
````

## Abstraction

Abstraction helps you to concentrate on the core functionality of an object without getting entangled in the intricacies of its implementation. Python's abc module (Abstract Base Classes) takes this a step further by helping you to define abstract base classes. These classes serve as blueprints for other classes, outlining essential methods that concrete subclasses must implement.

Example below shows how this works:

````python
from abc import ABC, abstractmethod

class Animal(ABC):  # Abstract base class
    @abstractmethod
    def make_sound(self):
        pass

class Dog(Animal):  # Concrete subclass
    def make_sound(self):
        return "Bark!"

class Mouse(Animal):  # Concrete subclass
    def make_sound(self):
        return "Squeak!"
````

In this example, Animal is an abstract base class. It declares an abstract method `make_sound()`, which doesn't have a concrete implementation. Any class inheriting from Animal (like Dog or Mouse) must provide its own implementation of `make_sound()`.

For instance, a `Shape` class may define methods such as `calculate_area()` and `calculate_perimeter()`, which are then implemented differently in subclasses like `Circle`, `Rectangle`, and `Triangle`. Polymorphism allows these methods to be called on any `Shape` object without needing to know its specific type.

Together, these concepts support the development of structured, reusable, and maintainable software capable of solving complex real-world problems.

# Forest Temperature Data

A group of students in an ecology class are collecting temperature data from a forest environment. To help organise and analyse the data, you will work with a Python class called `TemperatureData`.

The class stores the name of the location and a list of temperature readings. It also includes methods to calculate the average temperature and identify the highest and lowest recorded temperatures.


````python
class TemperatureData:
    # Constructor to create object
    def __init__(self, name, readings):
        self.name = name
        self.readings = readings

    # Calculate average temperatures
    def calculate_average_temp(self):
        return sum(self.readings) / len(self.readings)

    # Find highest temperature
    def find_high_temp(self):
        return max(self.readings)

    # Find lowest temperature
    def find_low_temp(self):
        return min(self.readings)
````
